```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gss_data = pd.read_stata('GSS7216_R1a.dta', convert_categoricals=False)
gss_data.head()
```

```python
gss_data.set_index('id')
gss_data.drop('id', 1, inplace=True)
gss_data.head()
```

```python
gss_data['age'].hist(bins=25)
plt.grid()
plt.locator_params(nbins=5);
```

```python
inc_age = gss_data[['realrinc','age']]
inc_age.head(10)
```

```python
inc_age = gss_data[['realrinc','age']].dropna()
inc_age.head(10)
```

```python
ax_list = inc_age.hist(bins=40, figsize=(8,3), xrot=45)
for ax in ax_list[0]:
    ax.locator_params(axis='x', nbins=6)
    ax.locator_params(axis='y', nbins=3)
```
